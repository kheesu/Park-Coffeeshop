```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
gn = pd.read_csv('data/final/gangnamm.csv')
ga = pd.read_csv('data/final/gwanakk.csv')
gj = pd.read_csv('data/final/gwangjin.csv')
sd = pd.read_csv('data/final/seongdong.csv')
sp = pd.read_csv('data/final/songpaa.csv')
ys = pd.read_csv('data/final/younsann.csv')
jr = pd.read_csv('data/final/jongrogu.csv')
jg = pd.read_csv('data/final/junggu.csv')
mp = pd.read_csv('data/final/mapogu.csv')

dobo = pd.read_csv('data/final/dobo.csv')
```

```python
df = pd.concat([gn, ga, gj, sd, sp, ys], ignore_index = 1)
df
```

```python
gn[gn['near'] == 0].iloc[:,1:5].mean(1)
```

```python
def rate(df):
    df['2019mean'] = df.iloc[:,1:4].mean(1)
    df['2020mean'] = df.iloc[:,5:9].mean(1)
    df['rate'] = (df['2020mean'] - df['2019mean']) / df['2019mean']
    
    df0 = df[df['near'] == 0]
    df1 = df[df['near'] == 1]
    
    return (df0, df1)
```

```python





ys0, ys1 = rate(gn)
```

```python
def qmean(df, near):
    out = []
    for i in range(1,9):
        out.append(df[df['near'] == near].iloc[:,i].mean())
    return out
```

```python
gn0 = qmean(gn, 0)
gn1 = qmean(gn, 1)

ga0 = qmean(ga, 0)
ga1 = qmean(ga, 1)

gj0 = qmean(gj, 0)
gj1 = qmean(gj, 1)

sd0 = qmean(sd, 0)
sd1 = qmean(sd, 1)

sp0 = qmean(sp, 0)
sp1 = qmean(sp, 1)

ys0 = qmean(ys, 0)
ys1 = qmean(ys, 1)

jr0 = qmean(jr, 0)
jr1 = qmean(jr, 1)

jg0 = qmean(jg, 0)
jg1 = qmean(jg, 1)

mp0 = qmean(mp, 0)
mp1 = qmean(mp, 1)
```

```python
plt.plot(gn0)
plt.plot(gn1)

print(sum(gn0)/len(gn0), sum(gn1)/len(gn1))

a, b = rate(gn)
stats.ttest_ind(a['rate'], b['rate'])
```
